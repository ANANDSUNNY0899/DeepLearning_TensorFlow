Sigmoid Function (Visualization)

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 200)
sig = sigmoid(z)

plt.plot(z, sig)
plt.title("Sigmois Function")
plt.xlabel("z")
plt.ylabel("sigmoid(z)")

plt.grid(True)
plt.show()
```

```python
def perceptron(z):
    return 1 if z >=0 else 0

step_output = [perceptron(val) for val in z]


plt.plot(z, step_output, label='perceptron Output')
plt.plot(z, sig, label='Sigmoid Output')
plt.legend()
plt.title("Perceptron Vs Sigmoid")
plt.show()
```
