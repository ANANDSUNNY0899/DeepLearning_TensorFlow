Sigmoid Function (Visualization)

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 200)
sig = sigmoid(z)

plt.plot(z, sig)
plt.title("Sigmois Function")
plt.xlabel("z")
plt.ylabel("sigmoid(z)")

plt.grid(True)
plt.show()
```
